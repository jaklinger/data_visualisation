```python
%matplotlib inline
from plotplot import boxflow
```

```python
data = {"Dinosaur" : {"Carrots" : 1,
                      "Peas" : 2,
                      "Chicken" : 3},
        "Cat" : {"Mouse" : 2,
                 "Chicken" : 1,
                 "Peas" : 0.05},
        "Dog" : {"Carrots" : 5,
                 "Chicken" : 3}}
```

```python
boxflow(data,threshold=0)
boxflow(data) # Cat - peas disappears

from matplotlib import pyplot as plt
fig, ax = plt.subplots(figsize=(6,10))
boxflow(data, ax=ax) 
```
