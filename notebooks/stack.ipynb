```python
%matplotlib inline
import pandas as pd
import plotplot as pp
```

```python
df = pd.DataFrame([{"animal":"dog","alive":"yes"},
                   {"animal":"dog","alive":"yes"},
                   {"animal":"dog","alive":"no"},
                   {"animal":"cat","alive":"no"},
                   {"animal":"cat","alive":"no"},
                   {"animal":"sheep","alive":"no"},
                   {"animal":"cat","alive":"yes"},
                   {"animal":"dog","alive":"maybe"},
                   {"animal":"dinosaur","alive":"yes"},
                   {"animal":"sheep","alive":"yes"},
                   {"animal":"cat","alive":"yes"},
                   {"animal":"dinosaur","alive":"no"}])
```

```python
pp.stack(df,x="animal",y="alive",cmap_name="Dark2")
pp.stack(df,x="animal",y="alive",y_order=["yes","no","maybe"])
pp.stack(df,x="animal",y="alive",fractions=True,legend=False)
```

```python
df = pd.DataFrame([{"animal":"dog","alive":"yes","weight":1.5},
                   {"animal":"dog","alive":"no","weight":3},
                   {"animal":"cat","alive":"no","weight":0.3},
                   {"animal":"cat","alive":"no","weight":1},
                   {"animal":"sheep","alive":"no","weight":0.8},
                   {"animal":"dinosaur","alive":"yes","weight":2.3}])

pp.stack(df,x="animal",y="alive",cmap_name="Dark2",weight="weight")
```
